```python
arr = [[' ' for _ in range(3)] for _ in range(3)]

def check(arr):
    for i in range(3):
        curr = arr[i][0]
        vcurr = arr[0][i]
        if curr != ' ':
            if all(arr[i][j] == curr for j in range(3)):
                return True
        if vcurr != ' ':
            if all(arr[j][i] == vcurr for j in range(3)):
                return True
    return False

def diag(arr):
    curr = arr[0][0]
    if curr != ' ' and all(arr[i][i] == curr for i in range(3)):
        return True
    curr = arr[0][2]
    if curr != ' ' and all(arr[i][2 - i] == curr for i in range(3)):
        return True
    return False

def displayer(arr):
    for it in arr:
        print('|'.join(it))
        print('-' * 5)
    print('-'*15)

k = 0
while k < 9:
    try:
        x = int(input("x = "))
        y = int(input("y = "))
    except ValueError:
        print("Invalid input. Enter numbers only.")
        continue

    if x >= 3 or y >= 3 or x < 0 or y < 0:
        print("Coordinates should be between 0 and 2. Try again.")
        continue
    
    if arr[x][y] == ' ':
        arr[x][y] = 'X' if k % 2 == 0 else 'O'
        k += 1
        displayer(arr)
        if check(arr) or diag(arr):
            print("Winner found!")
            print(arr[x][y])
            break
    else:
        print("Cell already taken. Try again.")

if k == 9:
    print("Game tie!")
```
